```python
import json
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure(num=1,figsize=(21,16))
plt.rc('font', size=14)

ax=fig.add_subplot(3,3,1)
path1 = "/root/autodl-tmp/llm/output/turns_20_budget=25_purchase_approx.json"
path2 = "/root/autodl-tmp/llm/output/turns_20_budget=30_purchase_approx.json"
path3 = "/root/autodl-tmp/llm/output/turns_20_budget=35_purchase_approx.json"
path4 = "/root/autodl-tmp/llm/output/turns_20_budget=40_purchase_approx.json"
with open(path1, "r") as f:
    y1 = json.load(f)
with open(path2, "r") as g:
    y2 = json.load(g)
with open(path3, "r") as f1:
    y3 = json.load(f1)
with open(path4, "r") as f2:
    y4 = json.load(f2)
    


xs = np.arange(21)
ax.set_xlabel("purchase-turns")#添加x轴坐标标签，后面看来没必要会删除它，这里只是为了演示一下。
ax.set_ylabel("uncertainty")
ax.plot(xs, np.array(y1), linestyle='-', marker="o", label="budget=25")
ax.plot(xs, np.array(y2), linestyle='dotted', marker="+", label="budget=30")
ax.plot(xs, np.array(y3), linestyle="dashed", marker="v", label="budget=35")
ax.plot(xs, np.array(y4), linestyle="-", marker="^",label="budget=40")
ax.set_xticks(xs[::5])
ax.legend(loc='upper center') 
   
ax2=fig.add_subplot(3,3,2)
path1 = "/root/autodl-tmp/llm/output/turns_20_budget=25_author_approx.json"
path2 = "/root/autodl-tmp/llm/output/turns_20_budget=30_author_approx.json"
path3 = "/root/autodl-tmp/llm/output/turns_20_budget=35_author_approx.json"
path4 = "/root/autodl-tmp/llm/output/turns_20_budget=40_author_approx.json"
with open(path1, "r") as f:
    y1 = json.load(f)
with open(path2, "r") as g:
    y2 = json.load(g)
with open(path3, "r") as f1:
    y3 = json.load(f1)
with open(path4, "r") as f2:
    y4 = json.load(f2)
    


xs2 = np.arange(21)
ax2.set_xlabel("author-turns")#添加x轴坐标标签，后面看来没必要会删除它，这里只是为了演示一下。
ax2.plot(xs, np.array(y1), linestyle='-', marker="o", label="budget=25")
ax2.plot(xs, np.array(y2), linestyle='dotted', marker="+", label="budget=30")
ax2.plot(xs, np.array(y3), linestyle="dashed", marker="v", label="budget=35")
ax2.plot(xs, np.array(y4), linestyle="-", marker="^",label="budget=40")
ax2.set_xticks(xs2[::5])
ax2.legend(loc='upper center') 


ax3=fig.add_subplot(3,3,3)
path1 = "/root/autodl-tmp/llm/output/turns_20_budget=25_employee_approx.json"
path2 = "/root/autodl-tmp/llm/output/turns_20_budget=30_employee_approx.json"
path3 = "/root/autodl-tmp/llm/output/turns_20_budget=35_employee_approx.json"
path4 = "/root/autodl-tmp/llm/output/turns_20_budget=40_employee_approx.json"
with open(path1, "r") as f:
    y1 = json.load(f)
with open(path2, "r") as g:
    y2 = json.load(g)
with open(path3, "r") as f1:
    y3 = json.load(f1)
with open(path4, "r") as f2:
    y4 = json.load(f2)
    


xs3 = np.arange(21)
ax3.set_xlabel("employee-turns")#添加x轴坐标标签，后面看来没必要会删除它，这里只是为了演示一下。
ax3.plot(xs, np.array(y1), linestyle='-', marker="o", label="budget=25")
ax3.plot(xs, np.array(y2), linestyle='dotted', marker="+", label="budget=30")
ax3.plot(xs, np.array(y3), linestyle="dashed", marker="v", label="budget=35")
ax3.plot(xs, np.array(y4), linestyle="-", marker="^",label="budget=40")
ax3.set_xticks(xs3[::5])
ax3.legend(loc='upper center') 

plt.show()
```
