```python
import json
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure(num=1,figsize=(6,4))
ax=fig.add_subplot(111)
path1 = "/root/autodl-tmp/llm/output/turns_20_k=3_employee_approx.json"
path2 = "/root/autodl-tmp/llm/output/turns_20_k=3_employee_random.json"
with open(path1, "r") as f:
    y1 = json.load(f)
with open(path2, "r") as g:
    y2 = json.load(g)

xs = np.arange(21)
ax.set_xlabel("turns")#添加x轴坐标标签，后面看来没必要会删除它，这里只是为了演示一下。
ax.set_ylabel("uncertainty",fontsize=10)
ax.plot(xs, np.array(y1), linestyle='-', marker="o", label="approx")
ax.plot(xs, np.array(y2), linestyle='-', marker='o', label="random")
ax.set_xticks(xs)
plt.legend(loc='upper center')    
plt.show()
```
